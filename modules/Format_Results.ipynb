```python
import pandas as pd
```

```python
df = pd.read_csv('Resultados.csv', sep=';')
```

```python
df
```

```python
def format_decimal(row):
    return str(row).replace(",", "").replace(".", ",")
```

```python
values_to_change = ["FO_inicial", "FO_final", "DP", "tempo"]
for value in values_to_change:
    df[value] = df[value].apply(format_decimal)
```

```python
df.to_excel('res.xlsx', index=False)
```
